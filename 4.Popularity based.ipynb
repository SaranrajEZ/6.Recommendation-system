```python
import pandas as pd
import numpy as np
```

```python
i_cols = ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
          'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
          'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('ml-100k/u.item', sep='|', names=i_cols, encoding='latin-1')
```

```python
items.head()
```

```python
r_cols = ['user id', 'movie id', 'rating', 'timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, encoding='latin-1')
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
movie_ratings = pd.merge(items, ratings, on='movie id')
```

```python
movie_ratings.head()
```

```python
movie_stats = movie_ratings.groupby('movie title').agg({'rating': ['mean', 'count']})
movie_stats.head()
```

```python
movie_stats.columns = ['mean_rating', 'rating_count']
popular_movies = movie_stats.sort_values(by=['rating_count', 'mean_rating'], ascending=False)
```

```python
popular_movies.head()
```

```python
N = 10
top_n_popular_movies = popular_movies.head(N)

print("Top N Popular Movies:")
print(top_n_popular_movies)
```
